```python
import pandas as pd
import re
import csv
import random
import nltk
import math
import numpy as np
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfTransformer
from scipy.spatial.distance import cdist
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.model_selection import KFold, train_test_split, learning_curve
from sklearn.svm import LinearSVC
from sklearn.metrics import roc_auc_score, roc_curve, auc

def clean(review):
    example1 = BeautifulSoup(review,'lxml').get_text()
    letters_only = re.sub("[^a-zA-Z]"," ", example1)
    words = letters_only.lower().split()
    words = [w for w in words if not w in stopwords.words('english')]
    return (" ".join(words))    

def dist(X, i, j, distance_function='euclidean'):
    return (pairwise_distances(X[i].reshape(1, -1), X[j].reshape(1,-1), metric=distance_function))

def topk(X, k):
    pair_dist = np.empty([X.shape[0],X.shape[1]])
    pair_dist = euclidean_distances(X, X)               #using inbuilt function to calculate distances
    pair_dist[pair_dist==0]=np.nan
    print ("new pair_dist")
    print (pair_dist)
    i=0
    final_list = []
    while(i<k):
        s = np.unravel_index(np.nanargmin(pair_dist), pair_dist.shape)
        min_val = pair_dist[s[0]][s[1]]
        pair_dist[s[0]][s[1]] = np.nan
        pair_dist[s[1]][s[0]] = np.nan
        pair_dist[pair_dist==min_val] = np.nan
        print (tuple((s[0], s[1], min_val)))
        final_list.append(tuple((s[0], s[1], min_val)))
        i=i+1
    return (final_list)
    
def part3(X, labels):
    
    X_train, X_test, y_train_list, y_test = train_test_split(X, labels, test_size=0.2, random_state=0)

    random.seed(0)
    list_C = np.random.uniform(low=1e-4, high=1e4, size=(30,))
    best_ras = -10
    
    for c in list_C:
        kf = KFold(n_splits=5)
        svc = LinearSVC(C=c)
        ras = 0.0
        
        for train_index, test_index in kf.split(X_train, y_train_list):
            y_train1 = []
            y_test1 = []
            X_train1 = X_train[train_index]
            X_test1 = X_train[test_index]
            for i in train_index:
                y_train1.append(y_train_list[i])
            for j in test_index:
                y_test1.append(y_train_list[j])

            svc.fit(X_train1,y_train1)
            answer1 = svc.predict(X_test1)
            ras += roc_auc_score(y_test1,answer1)
        
        ras = ras / 5
        print ("for ", c, " roc_auc_score is: ", ras)
        if(ras>best_ras):
            best_ras = ras
            best_C = c
    
    print ("best roc_auc_score is: ", ras)
    print ("best c is: ", best_C)
    
    #retraining classifier with best C
    svc = LinearSVC(C=best_C)
    svc.fit(X_train,y_train_list)
    answer = svc.decision_function(X_test)
    
    # Compute fpr, tpr, thresholds and roc auc
    fpr, tpr, thresholds = roc_curve(y_test, answer)
    roc_auc = auc(fpr,tpr)

    # Plot ROC curve
    plt.figure()
    plt.plot(fpr, tpr, lw=2, label='ROC curve (area = %0.3f)' % roc_auc)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate or (1 - Specifity)')
    plt.ylabel('True Positive Rate or (Sensitivity)')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
    print ("done")
    
    return best_C

def main():
    print ("started")
    
    train = pd.read_csv("labeledTrainData.tsv", header=0, delimiter="\t", quoting=3)
    
    # Create an empty list and append the clean reviews one by one
    num_reviews = len(train["review"])
    clean_train_reviews = [] 
    
    
    #print ("Cleaning and parsing the train set movie reviews...\n")
    for i in range(0,num_reviews):
        if( (i+1) % 1000 == 0 ):
            print ("Review %d of %d\n" % (i+1, num_reviews))
        clean_review = clean( train["review"][i] )
        clean_train_reviews.append( clean_review )
    
######################################################## PART ONE ###############################################################

    vectorizer = CountVectorizer(analyzer="word",tokenizer=None,preprocessor=None,stop_words=None,max_features=5000)
    
    #X_count is train_data_features
    train_data_features = vectorizer.fit_transform(clean_train_reviews).toarray()
    X_counts = np.array(train_data_features)
    print ("X_count")
    print (X_counts)
    print (X_counts.shape)
    print (X_counts.sum())
    
    #making X_binary
    X_binary = np.array(train_data_features)
    X_binary[X_binary>0]=1
    print ("X_binary")
    print (X_binary)
    print (X_binary.shape)
    print (X_binary.sum())
    
    #making X_tfidf
    transformer = TfidfTransformer(smooth_idf=False)
    X_tfidf = transformer.fit_transform(X_counts).toarray()
    print ("X_tfidf")
    print (X_tfidf)
    print (X_tfidf.shape)
    print (X_tfidf.sum())
    
    #making X_binary_imbalance
    labels = pd.DataFrame(train["sentiment"],columns=["sentiment"])
    labels_ones = labels[labels['sentiment'] == 1].index.tolist()
    length=len(labels_ones)
    random.seed(0)
    sampled_ones = random.sample(labels_ones,math.ceil(0.75*length))
    X_binary_imbalance = np.delete(X_binary, sampled_ones, 0)
    print ("X_binary_imbalance")
    print (X_binary_imbalance)
    print (X_binary_imbalance.shape)
    print (X_binary_imbalance.sum())
    
    
    labels_list = labels["sentiment"].tolist()
    labels_binary_imbalance = np.delete(labels_list, sampled_ones, 0)

######################################################## PART TWO ###############################################################
    
    k=10
    
    print ("First")
    final_list = topk(X_counts,k)
    i=0
    while(i<k):
        print ("distance is: ", final_list[i][2])
        print (final_list[i][0], " ", train["review"][final_list[i][0]][:20], " ", train.loc[final_list[i][0]]["sentiment"])
        print (final_list[i][1], " ", train["review"][final_list[i][1]][:20], " ", train.loc[final_list[i][1]]["sentiment"])
        print (" ")
        i=i+1
    
    print ("Second")
    final_list = topk(X_binary,k)
    i=0
    while(i<k):
        print ("distance is: ", final_list[i][2])
        print (final_list[i][0], " ", train["review"][final_list[i][0]][:20], " ", train.loc[final_list[i][0]]["sentiment"])
        print (final_list[i][1], " ", train["review"][final_list[i][1]][:20], " ", train.loc[final_list[i][1]]["sentiment"])
        print (" ")
        i=i+1
    
    print ("Third")
    final_list = topk(X_tfidf,k)
    i=0
    while(i<k):
        print ("distance is: ", final_list[i][2])
        print (final_list[i][0], " ", train["review"][final_list[i][0]][:20], " ", train.loc[final_list[i][0]]["sentiment"])
        print (final_list[i][1], " ", train["review"][final_list[i][1]][:20], " ", train.loc[final_list[i][1]]["sentiment"])
        print (" ")
        i=i+1
    
    print ("Fourth")
    final_list = topk(X_binary_imbalance,k)
    i=0
    while(i<k):
        print ("distance is: ", final_list[i][2])
        print (final_list[i][0], " ", train["review"][final_list[i][0]][:20], " ", train.loc[final_list[i][0]]["sentiment"])
        print (final_list[i][1], " ", train["review"][final_list[i][1]][:20], " ", train.loc[final_list[i][1]]["sentiment"])
        print (" ")
        i=i+1
    
######################################################## PART THREE #############################################################
    
    test = pd.read_csv("testData.tsv", header=0, delimiter="\t", quoting=3)
    
    # Create an empty list and append the clean reviews one by one
    num_reviews = len(test["review"])
    clean_test_reviews = [] 

    print ("Cleaning and parsing the test set movie reviews...\n")
    for i in range(0,num_reviews):
        if( (i+1) % 1000 == 0 ):
            print ("Review %d of %d\n" % (i+1, num_reviews))
        clean_review = clean( test["review"][i] )
        clean_test_reviews.append( clean_review )

    # Get a bag of words for the test set, and convert to a numpy array
    test_data_features = vectorizer.transform(clean_test_reviews)
    test_data_features = test_data_features.toarray()
    
    X_test_counts = np.array(test_data_features)
    X_test_binary = np.array(test_data_features)
    X_test_binary[X_test_binary>0]=1
    transformer1 = TfidfTransformer(smooth_idf=False)
    X_test_tfidf = transformer1.fit_transform(X_test_counts).toarray()
    
    c = part3(X_counts,labels_list)
    svc = LinearSVC(C=c)
    svc.fit(X_counts,labels_list)
    answer = svc.predict(X_test_counts)
    output = pd.DataFrame( data={"id":test["id"], "sentiment":answer} )
    output.to_csv( "X_counts_prediction.csv", index=False, quoting=3 )
    
    c = part3(X_binary,labels_list)
    svc = LinearSVC(C=c)
    svc.fit(X_binary,labels_list)
    answer = svc.predict(X_test_binary)
    output = pd.DataFrame( data={"id":test["id"], "sentiment":answer} )
    output.to_csv( "X_binary_prediction.csv", index=False, quoting=3 )
    
    c = part3(X_tfidf,labels_list)
    svc = LinearSVC(C=c)
    svc.fit(X_tfidf,labels_list)
    answer = svc.predict(X_test_tfidf)
    output = pd.DataFrame( data={"id":test["id"], "sentiment":answer} )
    output.to_csv("X_tfidf_prediction.csv", index=False, quoting=3 )
    
    c = part3(X_binary_imbalance,labels_binary_imbalance)
    svc = LinearSVC(C=c)
    svc.fit(X_binary_imbalance, labels_binary_imbalance)
    answer = svc.predict(X_test_binary)
    output = pd.DataFrame( data={"id":test["id"], "sentiment":answer} )
    output.to_csv( "X_imbalance_prediction.csv", index=False, quoting=3 )
    
######################################################## PART FOUR ##############################################################
    
    sizes = [100, 500, 1000, 2000, 3000, 4000, 5000, 7500, 10000, 15000, 20000]
    train_sizes, train_scores, valid_scores = learning_curve(LogisticRegression(), X_counts, labels_list, train_sizes=sizes, cv=5)
    print ("part 4")
    print (train_sizes)
    print (train_scores)
    print (valid_scores)
    train_score_mean = np.mean(train_scores, axis = 1)
    test_score_mean = np.mean(valid_scores, axis = 1)
    plt.plot(train_sizes, train_score_mean, 'o-', color="r",label="Training score")
    plt.plot(train_sizes, test_score_mean, 'o-', color="g",label="Cross-validation score")
    plt.legend(loc="lower right")
    plt.show()
    
if __name__ == "__main__":
    main()
```

###################################################### PART ONE #######################################################

X_count: This matrix contains the raw word counts of the input 'labeledTrainData.csv' after removing the punctuations and making it clean. The CountVectorizer has calculated the top 5000 features and then uses them to vectorize each row of the input text file. Thus, the resultant matrix is (number of reviews x 5000) dimensions. This matrix is filled in the following way: the number in the matrix corresponds to the number of times that word occurs in the line of text.
 
X_binary: This matrix just modifies the X_counts matrix by making elements to 0 and 1. If the number in the X_counts matrix is non-zero then it is replaced by a 1 else 0.
 
X_tfidf: This matrix basically performs a TfIdf transformation on the X_counts matrix. TfIdf stands for term frequency inverse document frequency which is a common term weighting technique. It works by scaling down the importance of commonly occuring words which are empirically less informative than those features which occur less times.
 
X_binary_imbalance: This matrix basically contains 75% less number of reviews which are labeled 1 as compared to X_counts. This is done in order to see the effect of imbalanced data wherein there exists more samples of one class than the other.

###################################################### PART TWO #######################################################

I have written the function dist() which uses the inbuilt function pairwise_distances to calculate the similarity between two rows of a matrix which are given as input based on any kind of distance metric such as Euclidean or Manhattan distance.

The topk() function calculates the top k=10 closest reviews in each of the 4 matrices which are passed to it. If there are many reviews having the same distances, I have chosen just one thus, my top 10 closest reviews are all having different distances. 

As can be seen from the output, the first few reviews in X_counts and X_binary are exactly the same and then they start differing as k approaches 10. This is because, there are many reviews which are almost the same and thus their distances are extremely small. Moreover, since I am printing just the first 20 letters, those mostly tend to be the same. 

X_tfidf show most of the reviews as the same since the distances between reviews are even smaller going into ranges of e-08.

X_binary_imbalance, on the otherhand, show a lot of differences. This could mainly be because of the imbalance in the numbers of reviews per class thus skewing the similarities.

Over all the matrices, the pairs given by X_counts and X_binary are almost the same but the other 2 matrices are very different. 

###################################################### PART THREE #####################################################

The main aim of this part was to find the best training parameter 'C' for a linear SVM by performing 5-fold cross validation on the each of the 4 design matrices. 'C' basically tells the classifier to what extent misclassifying should be avoided. A larger C means less allowance to misclassifying thus making a small margin SVM and vice versa. Cross validation on the training set is used to make the classifier more robust to overfitting and working better on the test data.

I found out the best C's by calculating the ROC-AUC values for each matrix. AUC stands for area under the curve and it determines how well a classifier is performing. The more the area, the better. This part was to train a logistic regression classifier on X_counts matrix and see the learning curve i.e. how it performs with different train data sizes. The plot of training score and cross validation score is seen in the above cell. The graphs are a plot of prediction score vs training data size and can help to decide how much we benefit from adding more training data and whether the estimator suffers more from a variance error or a bias error.

The plot for the learning curve is as shown below:

It can be seen that the training score is initially very high and as more number of training examples are considered, it starts reducing slightly. The cross validation score is initially very low but as number of examples increases, it increases rapidly. This is exactly the advantage of cross validating before applying the classifier on the test data. By increasing the number of training examples, the classifier can view larger variety of data and can classify better thus will have a better prediction accuracy on unseen test data. If the training score is much greater than the validation score for the maximum number of training samples, adding more training samples will most likely increase generalization.
